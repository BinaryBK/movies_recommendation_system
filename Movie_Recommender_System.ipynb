```python
import pandas as pd
import numpy as np
```

```python
#reading required dataset

movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# merging two dataframes on common column(title) basis as they contain data about the same movies

movies = movies.merge(credits, on = 'title')
movies.shape
```

```python
movies.head(1)
```

```python
#data cleaning : keeping only required columns as per our model requirements...

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(1)
```

```python
#checking for null rows in data 

movies.isnull().sum() #null rows 
```

```python
#checking for duplicate rows in data

movies.duplicated().sum()
```

```python
#removing null & duplicate values (if any)

movies.dropna(inplace=True)
movies.drop_duplicates()
```

```python
movies['genres']
#preprocessing :  formatting the data of columns in proper way for future use in model development
```

```python
#function for formatting
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
#defining a function to fetch first 3 Actors from the Cast

def fetch_3Actors(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter >= 3:
            break
        else:
            L.append(i['name'])
            counter += 1
    return L            
```

```python
movies['cast'] = movies['cast'].apply(fetch_3Actors)
```

```python
movies['cast']
```

```python
#defining a function to filter Director from Crew...

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['crew']
```

```python
movies['overview'][0]
```

```python
#changing overview datatype from string to list for Future Concatenation
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
#removing space from values so that every value will create unique and relavant tags for perfect recommendation

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#creating a column named 'tags' : main for recommendation

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
final_df = movies[['movie_id', 'title', 'tags']]
```

```python
#converting the tags as List to tags as string
final_df['tags'] = final_df['tags'].apply(lambda x:" ".join(x))
```

```python
#converting tags as string to lowercase which recommended always
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
```

```python
#final Dataframe for further model development
final_df.head()
```

```python
#now we apply stemming to transfrom [['love', 'loved', 'loves', 'loving']] to ----> 'love'
#this is done to ensure same thing shouldnt be appear as multiple feature of corpus

#we import a natural language processing library for this purpose

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
```

```python
final_df['tags'][0]
```

```python
final_df['tags'][1]
```

```python
#importing libraries for vectorization of tags as text
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors
```

```python
# bag of words : 5000 most repeated or common words 
feature_names = cv.get_feature_names_out()
pd.set_option('display.max_rows', None)
pd.Series(feature_names)
```

```python
#now we will calculate the Cosine distance of each movie vector with all other vector
#not Euclidean distance because it fails as the dimensions increases

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
#Function for recommendation 

def recommend(movie):
    movie_index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
